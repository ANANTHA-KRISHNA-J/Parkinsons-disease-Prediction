```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,classification_report
from sklearn.model_selection import train_test_split as tts
```

```python
df = pd.read_csv('parkinsons.csv').dropna()
df.tail(10)
```

```python
df.describe()
```

```python
df.info()
```

```python
X = df.drop(columns=['name','status'],axis=1)
y = df.status
```

```python
X.head()
```

```python
df['status'].value_counts()
```

```python
df.dtypes
```

```python
ss = StandardScaler()
X = ss.fit_transform(X)
```

```python
X[2]
```

```python
X_train,X_test,y_train,y_test = tts(X,y,test_size=0.25,random_state=6,stratify=y)
```

```python
print(X.shape , X_train.shape , X_test.shape , y_test.shape)
```

```python
model = SVC(kernel='sigmoid')
model.fit(X_train,y_train)
```

```python
pred_y = model.predict(X_test)
acc = accuracy_score(y_test,pred_y)
```

```python
print('Accuracy',acc)
```

```python
print("Classification report ",classification_report(y_test,pred_y))
```
